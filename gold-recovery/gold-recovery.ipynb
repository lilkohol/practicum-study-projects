### Предобработка данных

#### Анализ экстремальных значений в данных

Вернёмся к рассмотрению нулевых концентраций металлов в наших данных. Нам необходимо разобраться, являются нулевые значения тенденцией или ошибками, и почему так происходит.

```python
# добавим в выборку train столбцы с суммарными концентрациями трёх металлов, на этапе каждой очистки сырья 

# начальные значения 
col_list_1 = ['rougher.input.feed_ag', 'rougher.input.feed_pb', 'rougher.input.feed_au']
train['rougher.input.feed_total'] = train[col_list_1].sum(axis=1)

# черновой концентрат
col_list_2 = ['rougher.output.concentrate_ag', 'rougher.output.concentrate_pb', 'rougher.output.concentrate_au' ]
train['rougher.output.concentrate_total'] = train[col_list_2].sum(axis=1)

# после первичной очистки
col_list_3 = ['primary_cleaner.output.concentrate_ag', 'primary_cleaner.output.concentrate_pb', \
              'primary_cleaner.output.concentrate_au' ]
train['primary_cleaner.output.concentrate_total'] = train[col_list_3].sum(axis=1)

# после финальной очистки
col_list_4= ['final.output.concentrate_ag', 'final.output.concentrate_pb', 'final.output.concentrate_au' ]
train['final.output.concentrate_total'] = train[col_list_4].sum(axis=1)
```

```python
# для дальнейшего построения диаграммы рассеяния, объеденим суммарные концентрации металлов и 
# каждого из двух целевых признаков в таблицы

sub_train_1 = train.loc[:,['rougher.input.feed_total','rougher.output.concentrate_total',\
                           'primary_cleaner.output.concentrate_total','final.output.concentrate_total',\
                           'rougher.output.recovery']]

sub_train_2 = train.loc[:,['rougher.input.feed_total','rougher.output.concentrate_total',\
                           'primary_cleaner.output.concentrate_total', 'final.output.concentrate_total',\
                           'final.output.recovery']]
```

```python
# построим матрицу диаграмм рассеяния
sns.set_palette("muted")
scatter_plot = pd.plotting.scatter_matrix(sub_train_1, figsize = (12,12))  
for ax in scatter_plot.flatten():
    ax.xaxis.label.set_rotation(45)
    ax.yaxis.label.set_rotation(0)
    ax.yaxis.label.set_ha('right')
ax.set_facecolor('white')
plt.figure(facecolor='white')
plt.show()
```

```python
scatter_plot = pd.plotting.scatter_matrix(sub_train_2, figsize = (12,12))  
for ax in scatter_plot.flatten():
    ax.xaxis.label.set_rotation(45)
    ax.yaxis.label.set_rotation(0)
    ax.yaxis.label.set_ha('right')
plt.show()
```

Видим, что нулевые значения являются не тенденцией а исключением и значит ошибкой. Значит будем от них избавляться

```python
# найдём все строки с нулевым значением общей концентрации на каждом этапе
roughget_input_zero = train.loc[train['rougher.input.feed_total']==0]
rougher_out_zero = train.loc[train['rougher.output.concentrate_total']==0]
prime_out_zero = train.loc[train['primary_cleaner.output.concentrate_total']==0]
final_out_zero = train.loc[train['final.output.concentrate_total']==0]

print('Количество строк с нулевым значением общей концентрации металлов на этапе:')
print(f'первичной руды {len(roughget_input_zero)}')
print(f'после флотации {len(rougher_out_zero)}')
print(f'после первичной очистки {len(prime_out_zero)}')
print(f'в результате всего процесса {len(final_out_zero)}')
```

Метрикой для наших моделей будет симметричная средняя абсолютная процентная ошибка - sMAPE.

```python
# напишем функцию метрики, по которой мы будем оценивать успешность модели
def final_smape(real, predicted):
    real = real.to_numpy()
    smape = abs(0.25* (200/len(predicted[:,0]) * sum(abs(real[:,0] - predicted[:,0]) / \
                                                     (abs(real[:,0]) + abs(predicted[:,0])))) + \
            0.75* (200/len(predicted[:,1]) * sum(abs(real[:,1] - predicted[:,1]) / \
                                                 (abs(real[:,1]) + abs(predicted[:,1])))))
    return smape
```

```python
# sMAPE, если мы будем считать для каждого таргета отдельно
def smape_single_target(real, predicted):
    real = real.to_numpy()
    smape =200/len(predicted) * sum(abs(real[:,0] - predicted[:,0]) / \
                                                     (abs(real[:,0]) + abs(predicted[:,0])))
    return smape
```
